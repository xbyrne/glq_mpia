```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
band_names = ['g','r','i','z','Y','J','K','W1','W2']
photometry_df = pd.read_csv('../../selecting_data/objs_7102.csv',
                            index_col=0)
sylt_ids = np.load('../../contrastive_learning/sylt_ids.npz')['ids']
photometry_df.loc[sylt_ids]
```

```python
photometry_df['J_mag']
```

```python
file_text = '' # String to be written to .in file
for idd in tqdm(sylt_ids):
    file_text += str(idd)[-9:]
    for band in band_names:
        mag = photometry_df.at[idd, f"{band}_mag"]
        magerr = photometry_df.at[idd, f"{band}_magerr"]
        if mag == 99.: # Non-detection
            mag = -99
            magerr = -99
        file_text += f' {mag} {magerr}'
    file_text += '\n'
print(file_text)
```

```python
f = open('lephare_input.in', 'w')
f.write(file_text)
f.close()
```
